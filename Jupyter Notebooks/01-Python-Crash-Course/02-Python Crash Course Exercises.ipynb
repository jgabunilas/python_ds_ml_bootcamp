** Create a function that grabs the email website domain from a string in the form: **

    user@domain.com
    
**So for example, passing "user@domain.com" would return: domain.com**

```python
def domainGet(email):
  # split the string on the @ sign, creating a list in which the website domain will be the last item
  domain = email.split('@')
  # Return the last item in the list
  return domain[-1]
```

```python
print(domainGet('user@domain.com'))
```

```python
domainGet('user@domain.com')
```

** Create a function that counts the number of times the word "dog" occurs in a string. Again ignore edge cases. **

We'll solve this using the `str.count()` method.

```python
def countDog(input_str):
  return input_str.count('dog')
```
