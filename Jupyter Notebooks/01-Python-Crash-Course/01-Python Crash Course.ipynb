## Logic Operators

Logic operatos allow you to combine multiple comparison operators. Some basic rules:
- When using the `and` operator, ALL comparison statements must resolve to True in order for the whole statement to resolve to True
- When using the `or` operator, only one statement must resolve to True in order for the whole statement ot resolve to True

```python
(1 > 2) and (2 < 3)
```

```python
(1 > 2) or (2 < 3)
```

```python
(1 == 2) or (2 == 3) or (4 == 4)
```

It is always recommended to use parentheses with logic operators as it makes things more readable and helps prevent errors.

## list comprehension

List comprehension is a concise way of constructing a list based on inputs from a different list. You never need to use it, but it can be more convenient that doing *for* loops.

Here is how we generate a list of squares using a *for* loop.

```python
x = [1,2,3,4]
```

```python
out = []
for item in x:
    out.append(item**2)
print(out)
```

This list comprehension accomplishes the same task with less code, though it is less human readable.

```python
[item**2 for item in x]
```

## functions

Functions are groups of code that perform a particular set of actions. A determining aspect of functions is that they can be called and used as many times of you wish, making them great for operations that need to be repeated.

Functions defined with `def` and can be defined with parameters. You can also add **docstrings** to functions where you can explain what the function does.

## lambda expressions

Lambda expressions, also known as anonymous functions, are a quick way to use a function only once. It does not assigned to a name, and is not callable.

Consider this function that multiplies the supplied value by 2.

```python
def times2(var):
    return var*2
```

```python
times2(2)
```

This lambda expression does the same thing. We'll go into more detail about this later on when we use it later in the course.

```python
lambda var: var*2
```

## map and filter

The `map()` function creates an iterator that computes the given function for each of the iterabes.

```python
seq = [1,2,3,4,5]
```

This example will use the `times2()` function defined above.

```python
map(times2,seq)
```

The mapped object needs to be cast into a list.

```python
list(map(times2,seq))
```

Can also be done with lambda functions.

```python
list(map(lambda var: var*2,seq))
```

The `filter()` function creates an iterator that returns all objects in the iterable for which the function resolves to True

```python
filter(lambda item: item%2 == 0,seq)
```

```python
list(filter(lambda item: item%2 == 0,seq))
```
